```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Spill_incidents.csv')
```

```python
df.head(10)
```

```python
df.describe()
```

### Converting the Date columns to datatime

```python
df[['Spill Date', 'Received Date', 'Close Date']] = pd.to_datetime(df[['Spill Date', 'Received Date', 'Close Date']].stack()).unstack()
```

```python
df.head()
```

```python
oldest_time = min(df['Spill Date'])
recent = max(df['Spill Date'])
print(f"Oldest: {oldest_time}, Most recent: {recent}")
```

### Creating a Year column in the dataset

```python
df['Year'] = df['Spill Date'].dt.year
```

```python
df['Year'].value_counts().sort_index().plot.bar(figsize=(100,50))
```

### There isn't enough data from the early years (as expected), so I want to create a dataframe that only covers from 1990 to 2021. 


```python
years = list()
for i in range(1990, 2022):
    years.append(float(i))
    i += 1
```

```python
df_years = pd.DataFrame()

for year in years:
    df2 = df[df['Year'] == year]
    df_years = df_years.append(df2, ignore_index = True)
```

```python
df_years.head()
```

### Spill Incidents from 1990 to 2021

```python
Years_series = df_years['Year'].value_counts().sort_index()
plt.title("Spill Incidents from 1990 to 2021")
Years_series.plot.bar(figsize=(15,10))
```

### Monthly distribution of spills (from 1990 to 2021)

```python
df_years['month'] = df['Spill Date'].dt.month
```

```python
def month_update(month):
    if month == 1.0: return 'January'
    if month == 2.0: return 'February'
    if month == 3.0: return 'March'
    if month == 4.0: return 'April'
    if month == 5.0: return 'May'
    if month == 6.0: return 'June'
    if month == 7.0: return 'July'
    if month == 8.0: return 'August'
    if month == 9.0: return 'September'
    if month == 10.0: return 'October'
    if month == 11.0: return 'November'
    if month == 12.0: return 'December'

    
```

### Creating and categorizing a month column in the dataset

```python
df_years['month'] = df_years['month'].apply(lambda x: month_update(x))
```

```python
months = ["January", "February", "March", "April", "May", "June", 
          "July", "August", "September", "October", "November", "December"]

df_years['month'] = pd.Categorical(df_years['month'], categories=months, ordered=True)
```

### How often does spillage occur each month? (from 1990 to 2021)

```python
month_series = df_years['month'].value_counts().sort_index()
plt.title("Spill incident frequency by month (from 1990 to 2021)")
month_series.plot.bar(figsize=(10,10))
```
